```python
##Importing libraries
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
from pylab import rcParams
import urllib
import sklearn
from sklearn.neighbors import KNeighborsClassifier               ##KNN
from sklearn.neighbors.nearest_centroid import NearestCentroid   ##Rochio
from sklearn.naive_bayes import BernoulliNB                      ##Naive Bayes
from sklearn import neighbors 
from sklearn import preprocessing
from sklearn.cross_validation import train_test_split
from sklearn import metrics

##Upload input data file
address='C:/Users/HP/Desktop/dbworld_bodies.csv'
#address='C:/Users/madiha/Desktop/dbworld_bodies.csv'
dbworld = pd.read_csv(address).values

X = dbworld[:,0:4702] #Data columns
Y = dbworld[:,-1] 


#Splitting data into train test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0) 

#################
## KNN Classifier
clf = neighbors.KNeighborsClassifier()
clf.fit(X_train, Y_train)
print(clf)

Y_Expect= Y_test      #Y test was original expected value
Y_Pred = clf.predict(X_test)    #Predicting Class from test data X

#Print KNN results
print("KNN Results:")
print(metrics.classification_report(Y_Expect, Y_Pred))


####################
## Rocchio Classifier
clf1 = NearestCentroid()
clf1.fit(X_train, Y_train)
NearestCentroid(metric='euclidean', shrink_threshold=None)
print(clf1)

Y_Expect1 = Y_test
Y_Pred1 = clf1.predict(X_test)
print("Rocchio Results:")
print(metrics.classification_report(Y_Expect1, Y_Pred1))


####################
## Naive Bayes

clf2 = BernoulliNB()

clf2.fit(X_train, Y_train)
BernoulliNB(alpha = 1.0, binarize = 0.0, class_prior = None, fit_prior = True)
print(clf2)

Y_Expect2 = Y_test
Y_Pred2 = clf2.predict(X_test)
print("Naive Bayes Results:")
print(metrics.classification_report(Y_Expect2, Y_Pred2))
```
